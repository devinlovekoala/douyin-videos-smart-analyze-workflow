# 智能短视频内容理解与分发系统

基于字节开源 Seed1.5-VL 多模态模型的短视频内容理解系统

## 核心应用场景

### 🎯 内容理解标签系统
利用多模态AI技术，构建12+维度的结构化内容标签体系，支持精准的内容分类和检索

### 📊 个性化分发
- **传播特征分析**: 评估视频的话题性、病毒性和用户粘性
- **分发场景匹配**: 智能匹配推荐场景（首页/关注页/话题页等）
- **受众定位**: 精准识别目标用户群体

### 🎨 画风识别
- **视觉风格分类**: 识别写实、卡通、复古、赛博朋克等多种风格
- **色彩分析**: 提取色调、饱和度、对比度等特征
- **制作水平评估**: 区分专业制作与个人创作

### 🛡️ 社区氛围治理
- **内容健康度评估**: 检测不当内容和争议话题
- **价值导向分析**: 评估内容的社会影响
- **氛围建设**: 识别正向/负向的社区影响

## 分析流程

1. **视频扫描** → 识别待分析视频
2. **5维度深度分析** → 内容理解、画风识别、氛围分析、传播特征、多模态特征
3. **智能标签生成** → 12+维度结构化标签
4. **价值评估** → 综合评分和推荐策略
5. **报告输出** → 完整分析报告

---

将屏幕录制的视频放入 `download_videos` 目录，依次执行下面的代码块开始分析

```python
# 导入必要的库
import json
from pathlib import Path
from video_analyzer import VideoAnalyzer
from content_labeling_system import ContentLabelingSystem
from douyin_ecosystem_analyzer import DouyinEcosystemAnalyzer
```

## 步骤 1: 检查视频文件

扫描 `download_videos` 目录中的视频文件

```python
# 扫描视频目录（支持 mp4 和 webm 格式）
video_dir = Path("download_videos")
video_files = sorted([
    f for f in video_dir.glob("*") 
    if f.suffix.lower() in ['.mp4', '.webm', '.avi', '.mov']
])

print(f"发现 {len(video_files)} 个视频文件:")
for i, video_file in enumerate(video_files, 1):
    size_mb = video_file.stat().st_size / (1024 * 1024)
    print(f"  {i}. {video_file.name} ({size_mb:.2f} MB)")
```

## 步骤 2: 视频内容分析

使用 Seed1.5-VL 多模态模型分析视频内容

```python
# 初始化视频分析器
analyzer = VideoAnalyzer(api_key="a923c210-64ac-43db-b0b9-5122b11a6aa7")

# 结果目录
results_dir = Path("results")
results_dir.mkdir(parents=True, exist_ok=True)

# 批量分析视频（并发处理，支持多种分析类型）
video_analyses = analyzer.batch_process(
    video_dir="./download_videos",
    output_file=str(results_dir / "video_analysis_results.json"),
    analysis_types=("content_understanding", "visual_style", "quality", "tags"),
    max_workers=16,
)

print(f"✅ 完成 {len(video_analyses)} 个视频的分析")
```

## 步骤 3: 智能标签生成

基于视频分析结果生成多维度标签

```python
# 初始化标签生成系统
labeling_system = ContentLabelingSystem()

# 批量生成标签（复用步骤2的分析结果，无需重新调用模型）
video_labels = labeling_system.batch_label_videos(
    video_analyses,
    output_file=str(results_dir / "video_labels.json")
)

print(f"✅ 完成 {len(video_labels)} 个视频的标签生成")
```

## 步骤 4: 生态价值评估

评估视频在抖音生态中的传播潜力和价值

```python
# 初始化生态分析器
ecosystem_analyzer = DouyinEcosystemAnalyzer()

# 批量生态价值评估（复用标签和分析结果）
ecosystem_scores = ecosystem_analyzer.batch_ecosystem_scores(
    video_labels=video_labels,
    video_analyses=video_analyses,
    output_file=str(results_dir / "ecosystem_scores.json")
)

print(f"✅ 完成 {len(ecosystem_scores)} 个视频的生态价值评估")
```

## 步骤 5: 生成分析报告

汇总所有分析结果，生成综合报告

```python
# 生成综合报告
from collections import Counter

# 统计标签分布
all_tags = []
for label_data in video_labels.values():
    if "tags" in label_data:
        for tag_list in label_data["tags"].values():
            all_tags.extend(tag_list)

tag_distribution = Counter(all_tags).most_common(10)

# 按评分排序视频
sorted_videos = sorted(
    [(vid, data) for vid, data in ecosystem_scores.items() if "overall_score" in data],
    key=lambda x: x[1]["overall_score"],
    reverse=True
)

# 生成报告
report = {
    "summary": {
        "total_videos": len(video_files),
        "analyzed_videos": len(video_analyses),
        "labeled_videos": len(video_labels),
        "scored_videos": len(ecosystem_scores)
    },
    "top_videos": [
        {
            "video_id": vid,
            "score": data["overall_score"],
            "recommendation": data["recommendation"],
            "key_tags": list(video_labels.get(vid, {}).get("tags", {}).get("content_tags", []))[:3]
        }
        for vid, data in sorted_videos[:5]
    ],
    "statistics": {
        "avg_score": sum(s["overall_score"] for s in ecosystem_scores.values() if "overall_score" in s) / max(len([s for s in ecosystem_scores.values() if "overall_score" in s]), 1),
        "high_quality_count": sum(1 for s in ecosystem_scores.values() if s.get("overall_score", 0) >= 0.6),
        "top_tags": [{"tag": tag, "count": count} for tag, count in tag_distribution]
    }
}

# 保存报告
with open(results_dir / "analysis_report.json", "w", encoding="utf-8") as f:
    json.dump(report, f, ensure_ascii=False, indent=2)

# 显示报告摘要
print("=" * 60)
print("📊 分析报告摘要")
print("=" * 60)
print(f"\n总视频数: {report['summary']['total_videos']}")
print(f"分析完成: {report['summary']['analyzed_videos']}")
print(f"平均评分: {report['statistics']['avg_score']:.2f}")
print(f"高质量视频: {report['statistics']['high_quality_count']}")

print(f"\n🏆 排名前5的视频:")
for i, video in enumerate(report['top_videos'], 1):
    print(f"  {i}. {video['video_id']}")
    print(f"     评分: {video['score']:.2f} | {video['recommendation']}")
    if video['key_tags']:
        print(f"     标签: {', '.join(video['key_tags'])}")

print(f"\n🏷️ 热门标签 TOP 10:")
for tag_info in report['statistics']['top_tags']:
    print(f"  • {tag_info['tag']}: {tag_info['count']} 次")

print(f"\n✅ 所有结果已保存到 results/ 目录")
```
